```python
import json
```

```python
todouhuken = []
hiduke = []
data = {}
with open('Covid19JapanAll.json', encoding="utf-8") as f:
    df = json.load(f)
    data = df["itemList"]

print(df["itemList"][0:47])
for i in range(0, 47):
    todouhuken.append(data[i]["name_jp"])
print(todouhuken)

for i in range(0, len(data), 47):
    hiduke.append(data[i]["date"])
print(hiduke)
    
```

```python
data_dict = {}
for day_pref in reversed(data):
    today = day_pref["date"]
    if not(today in data_dict.keys()):
        data_dict[today] = {}
    data_dict[today][day_pref["name_jp"]] = {}
    if today == "2020-04-22":
        data_dict[today][day_pref["name_jp"]]["per_day"] = int(day_pref["npatients"])
        data_dict[today][day_pref["name_jp"]]["ruiseki"] = int(day_pref["npatients"])
    else:
        yesterday = hiduke[hiduke.index(today)+1]
        try:
            data_dict[today][day_pref["name_jp"]]["ruiseki"] = int(day_pref["npatients"])
            data_dict[today][day_pref["name_jp"]]["per_day"] = int(day_pref["npatients"])-int(data_dict[yesterday][day_pref["name_jp"]]["ruiseki"])
        except:
            print("error")
            break
print(data_dict['2020-04-30'])
```

```python
with open('newCovid19JapanAll.json', 'w') as f:
    json.dump(data_dict, f)
```
